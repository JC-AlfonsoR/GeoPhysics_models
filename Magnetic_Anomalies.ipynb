## Isolated Pole
$$F_p=(-p/r^2)r_1=(p/r^3)(-x\vec{i}-y\vec{j}+z_p\vec{k})$$
$$f=F_p\dot f_1=(p/r^3)(-xcos(I)+z_psin(I))$$

```python
z = 1.0
x = linspace(-4,4,100)
r = (x**2 + z**2)**0.5
I = deg2rad(linspace(0,90,4))
```

```python
for i in I:
    f = (-x*cos(i) + z*sin(i))/r**3
    plot(x,f,label='I='+str(rad2deg(i)))
grid()
legend(loc='best',framealpha=True)
xlabel(r'$\frac{x}{z_p}$',size=30)
ylabel(r'$\frac{F}{p/z_p^2}$',size=20)
title('Isolated Pole Anomaly')
```

## Sphere
Vertical Magnetization

```python
R = 1000 # Sphere radius [m]
dk = 1e-1 - 1e-4 # Magnetic suceptibility delta[SI]
x = linspace(-16e3,16e3,1000) # x domain
Mu0 = 4*pi*10e-7 # Magnetic Permeability [SI]
z = 1000 # Depth [m]

DBz = 1/3*Mu0*R**3*dk*(2*z**2-x**2)/((z**2+x**2)**(5/2))
DBz_n = DBz/max(abs(DBz))
```

```python
figure(figsize=(8,4))
plot(x/1000,DBz_n)
grid(); xlim([-5,5])
xlabel('x [m]'); ylabel(r'$\Delta B_z$ normalized')
```

## Horizontal Cylinder
Vertical Magnetization

```python
R = 1000 # Cylinder radius [m]
dk = 1e-1 - 1e-4 # Magnetic suceptibility delta[SI]
x = linspace(-16e3,16e3,1000) # x domain
Mu0 = 4*pi*10e-7 # Magnetic Permeability [SI]
z = 1000 # Depth [m]

DBz = 1/2*Mu0*R**2*dk*(z**2-x**2)/((z**2+x**2)**(2))
DBz_n = DBz/max(abs(DBz))
```
